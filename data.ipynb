```python
train_df = pd.read_csv('train_train_set.csv',delimiter='\t')
valid_df = pd.read_csv('validation_train_set.csv',delimiter='\t')
test_df = pd.read_csv('test_set.csv',delimiter='\t')
```

```python
df = pd.concat([train_df,valid_df,test_df])
```
